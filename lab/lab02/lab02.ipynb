# Part 1: Our Alphabet

In this course, unless otherwise noted, you can assume plaintext and ciphertext are characters from the English language. We will need a reliable and consistent way of converting these characters into an equivalent numerical form.

While there are more sophisticated and modern ways of converting characters to numbers, and vice versa, in this course we'll use a string called `LETTERS` to hold the characters of our alphabet, where the index of the character is the numerical representation of that character. This allows us to use string indexing, `LETTERS[...]` to quickly compute the characters that correspond to a particular numerical value.

```python
LETTERS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

For example, 
*  Since `LETTERS[0]` is `'A'`, $A \rightarrow 0$
*  Since `LETTERS[3]` is `'D'`, $D \rightarrow 3$

So, the characters `M`, `A`, `T`, and `H` could be converted to the equivalent numerical values of: `12`, `0`, `19`, `7`
