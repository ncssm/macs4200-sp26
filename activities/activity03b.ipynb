### String Methods to Consider Using
* `.upper()` - makes all letters uppercase
* `.lower()` - makes all letters lowercase
* `.title()` - capitalizes first letter of each word
* `.capitalize()` - capitalizes first letter of the string
* `.strip()` - removes extra spaces from ends
* `.split()` - breaks string into parts
* `.join()` - combines strings together
