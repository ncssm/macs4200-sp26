```python
def find_repeating_substrings(s):
    """
    Find all maximal substrings of length 4+ that appear more than once.
    A substring is excluded if it's fully contained within every occurrence
    of a longer repeating substring.

    Returns a dict mapping each repeating substring to a list of
    starting indices where it occurs.
    """
    n = len(s)
    all_repeats = {}

    for length in range(4, n):
        for i in range(n - length + 1):
            sub = s[i:i + length]
            if sub not in all_repeats:
                all_repeats[sub] = []
            all_repeats[sub].append(i)

    # Keep only substrings that actually repeat
    repeats = {sub: idx for sub, idx in all_repeats.items() if len(idx) > 1}

    # Filter out non-maximal substrings: remove a substring if every one of
    # its occurrences is contained within an occurrence of a longer repeat.
    to_remove = set()
    for sub, indices in repeats.items():
        for longer, longer_indices in repeats.items():
            if len(longer) <= len(sub) or longer == sub:
                continue
            # Check if every occurrence of `sub` falls inside an occurrence of `longer`
            if all(
                any(li <= si and si + len(sub) <= li + len(longer)
                    for li in longer_indices)
                for si in indices
            ):
                to_remove.add(sub)
                break

    return {sub: idx for sub, idx in repeats.items() if sub not in to_remove}
```
