## Helper Functions

The functions below are from Lab 07. Run the cell to load them.

```python
def get_bit(value, bit_index):
    bitmask = 1 << bit_index
    result = value & bitmask
    return result >> bit_index

def set_bit(value, bit_index):
    bitmask = 1 << bit_index
    result = value | bitmask
    return result 

def clear_bit(value, bit_index):
    bitmask = 1 << bit_index
    result = value & ~bitmask
    return result
```

The cell below contains the code to implement the 6-bit LFSR described in Activity 24.

```python
def lfsr6(seed, num_bits):
    current_state = seed
    output = 0
    
    for i in range(num_bits):
        
        output_bit = get_bit(current_state, 0)
        shifted_output = output << 1
        
        if output_bit == 1:
            output = set_bit(shifted_output, 0)
        else:
            output = clear_bit(shifted_output, 0)
        
        new_bit = get_bit(current_state, 2) ^ get_bit(current_state, 1) ^ get_bit(current_state, 0)
        
        shifted_current_state = current_state >> 1
        
        if new_bit == 1:
            current_state = set_bit(shifted_current_state, 5)
        else:
            current_state = clear_bit(shifted_current_state, 5)
    
    return output
```
