The crux of the security of Diffie Hellman is that it's very difficult to determine the exponent $a$ when all you have is the value of $g^a \pmod{p}$. For example, if you intercepted $g^a = 389,261,021$ from Alice it would be very difficult to determine the value of $a$ even if you know $g = 7$ and $p = 1,000,000,007$. You might have to try all values of a less than $p$, which could take a very long time! Let's explore how long it would take to "brute force" values of $a$, for different sized values of $p$.

The function `crack_diffie_hellman(g, p, A)` tries all possible values of $a$ between $1$ and $p$ until it finds the a value that satisfies $A \equiv g^a \pmod{p}$.

```python
def crack_diffie_hellman(g, p, A):
    for a in range(1,p):
        if pow(g,a,p) == A:
            return(a)
```

For example, the code below will try all possible values of $a$ (the integers between $1$ and $37$) until it determines which value satisfies $7 \equiv 5^a \pmod{37}$.

```python
crack_diffie_hellman(5, 37, 3)
```

Python has the ability to run code and time how long it takes to complete. This is done by using the `timeit` library. The following code will execute the command `crack_diffie_hellman(5, 37, 7)` 100 times in a batch, then run 5 rounds of batches, and finally return the average time, it took for it to find the value for $a$, when $7 \equiv 5^a \pmod{37}$. Be aware of the units on the output, as they could be microseconds, milliseconds, or just seconds, depending on how long it takes.

```python
%timeit -n10 -r100 crack_diffie_hellman(3, 37, 7)
```
