### Question 1

**Goal**: Print the numbers from 1 to 5

```python
counter = 1
while counter <= 5:
    print(counter)
```

### Question 2

**Goal**: Print each character in the string `word`, where each character is on its own line.

```python
word = "CRYPTO"
for i in range(1, len(word)):
    print(word[i])
```

### Question 3

**Goal**: Print out numbers in the pattern: 1, 2, 4, 8, 16

```python
number = 1
while number <= 16:
    number = number * 2
    print(number)
```

### Question 4

### Question 5

**Goal**: Sum numbers from 1 to 5

```python
sum = 0
for number in range(1, 6):
    sum = number
print(sum)
```

### Question 6

**Goal**: Count how many times character `A` appears in the string `word`, and print it once.

```python
word = "PANAMA"
count = 0
for letter in word:
    count = 0
    if letter == "A":
        count = count + 1
print(count)
```

### Question 7

**Goal**: Encipher the string `message` using the Caesar cipher with key value stored to `key`.

```python
LETTERS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
message = 'HELLO' 
key = 3 
result = '' 

for letter in message: 
    result = LETTERS[(LETTERS.index(letter) + key) % len(LETTERS)] 

print(result)
```
