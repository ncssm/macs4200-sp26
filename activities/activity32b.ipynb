# Activity 32: Exploring RSA


This activity will show the actual tools modern computers use to generate RSA keypairs, and have you verify the mathematics that they use is the same they we've learned in class.



````python
from math import gcd
````

````python
def phi(n):
    total = 0
    for i in range(1,n):
        if gcd(n, i) == 1:
            total += 1

    return total
````

Open a terminal in this folder, and run the following command

```bash
openssl genrsa -out private_key.pem 512
```

This will generate a private key named `private_key.pem` with a 512-bit modulus (n).

You can see the contents of the key by running:

```bash
openssl rsa -in private_key.pem -text -noout
```

But, the numerical values are shown in hexadecimal instead of decimal. The code cell below will show you the values of your private key in decimal format. You can see the essential public information is stored to `pub.n` and `pub.e` and the essential private numbers to `nums.d`, `nums.p`, `nums.q`. The other numbers `nums.dmp1`, `nums.dmq1`, and `nums.iqmp` are not technically required to do RSA, but there's some nice mathematics that can be done with these numbers to speed up the exponentiation process using the Chinese Remainder Theorem, which makes it about 4x faster. We won't use those numbers in this course.

````python
from cryptography.hazmat.primitives import serialization

with open("private_key.pem", "rb") as f:
    key = serialization.load_pem_private_key(f.read(), password=None)

nums = key.private_numbers()
pub = nums.public_numbers

print(f"n (modulus)          = {pub.n}")
print(f"e (public exponent)  = {pub.e}")
print(f"d (private exponent) = {nums.d}")
print(f"p (prime1)           = {nums.p}")
print(f"q (prime2)           = {nums.q}")
print(f"dp (d mod p-1)       = {nums.dmp1}")
print(f"dq (d mod q-1)       = {nums.dmq1}")
print(f"qinv (q^-1 mod p)    = {nums.iqmp}")
````

If you needed a public key to send to someone to use, all the information is already in the private key. You can have your computer package up the information using the following terminal command:

```bash
openssl pkey -in private_key.pem -pubout -out public_key.pem
```

````python
with open("public_key.pem", "rb") as f:
    pub_key = serialization.load_pem_public_key(f.read())

pub_nums = pub_key.public_numbers()
print(f"n (modulus)         = {pub_nums.n}")
print(f"e (public exponent) = {pub_nums.e}")
````
